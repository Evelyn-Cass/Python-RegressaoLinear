```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
try:
    dados = np.loadtxt('dataset.csv', delimiter=',', skiprows=1)
except FileNotFoundError:
    print("Arquivo 'dataset.csv' não encontrado. Certifique-se de que o arquivo está no diretório correto.")
    exit()  
```

```python
print(dados)
```

```python
x = dados[:, 0]  # Coluna 1: x
y = dados[:, 1]  # Coluna 2: y

print("x:", x)
print("y:", y)

print("Núumero de amostras:", x.shape[0]) 
```

```python
x = x.reshape(-1, 1)  # Transformar x em um array 2D (n_samples, n_features)
y = y.reshape(-1, 1)  # Transformar y em um array 2D (n_samples, n_targets)

print("x (reshaped):\n", x)
print("y (reshaped):\n", y)
```

```python
print("Número de características:", x.shape[1])
print("Número de saídas:", y.shape[1])
print("Número total de elemento de X:", x.size)
print("Número total de elemento de y:", y.size)
print("Tipo de dados em x:", x.dtype)
print("Tipo de dados em y:", y.dtype)
print("Tamanho em bytes de cada elemento em x:", x.itemsize)
print("Tamanho em bytes de cada elemento em y:", y.itemsize)
print("Número total de bytes consumidos por x:", x.nbytes)
print("Número total de bytes consumidos por y:", y.nbytes)
print("Número de dimensões de x:", x.ndim)
print("Número de dimensões de y:", y.ndim)
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

print("Média de x:", x_mean)
print("Média de y:", y_mean)
```
